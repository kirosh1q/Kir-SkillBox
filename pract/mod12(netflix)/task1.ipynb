```python
import pandas as pd
import sqlite3

conn = sqlite3.connect('netflix.sqlite')
cursor = conn.cursor()

```

Создал таблицу Actors

```python
cursor.execute('''CREATE TABLE Actors (
                    id INTEGER PRIMARY KEY,
                    actor_name TEXT
                )''')
```

Перенос id и cast

```python
cursor.execute('''INSERT INTO Actors (id, actor_name)
                  SELECT show_id, [cast]
                  FROM netflix_titles''')
```

```python
pd.read_sql("SELECT * FROM Actors", conn)
```

Создание таблицы актеры-фильм

```python
cursor.execute('''CREATE TABLE actor_films (
                    id INTEGER,
                    actors TEXT,
                    film_title TEXT,
                    FOREIGN KEY (id) REFERENCES Actors(id)
                )''')
```

Заполнение таблицы

```python
cursor.execute('''INSERT INTO actor_films (id, actors, film_title)
                  SELECT nt.show_id, ac.actor_name, nt.title
                  FROM netflix_titles nt
                  INNER JOIN Actors ac ON nt.show_id = ac.id''')
```

```python
pd.read_sql("SELECT * FROM actor_films", conn)
```

```python
conn.commit()
conn.close()
```
